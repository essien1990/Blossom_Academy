```python
# import pandas
import pandas as pd
import pandas as pd
# read or import CSV file and view the data
bl = pd.read_csv('billboard.csv')
bl
```

```python
# use melt to categorize the colmuns items under each specific column name and view the changed data
melted_bl = pd.melt(bl, id_vars=['artiste','track', 'genre','date.entered', 'date.peaked'], var_name='week'
                    , value_name='rank')
melted_bl
```

```python
melted_bl
```

```python
# cleaning unnecessary row and view the change
melted_bl = melted_bl.dropna()
```

```python
melted_bl
```

```python
#extrat the weeks and convert them to integer
melted_bl['week'] = melted_bl['week'].str.extract('(\d+)').astype(int)
```

```python
melted_bl
```

```python
# substract the peaked date from the date entered to the see the number of days per ranking of the songs
melted_bl['date'] = pd.to_datetime(melted_bl['date.peaked']) - pd.to_datetime(melted_bl['date.entered'])
```

```python
# extracted date in days
melted_bl['date'].dt.days
```
