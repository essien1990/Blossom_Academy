```python
# import boto3 library
import boto3
import zipfile
# import botocore library for error handling
import botocore

s3 = boto3.resource('s3')

try:
    s3.Bucket('blossom-data-engs').download_file('free-7-million-company-dataset.zip', '/Users/nanayaw/Desktop/blossom/blossomenv/s3project.zip')
except botocore.exceptions.ClientError as e:
    if e.response['Error']['Code']=='404':
        print('The file does not exist')
    else:
        raise

with zipfile.ZipFile('s3project.zip','r') as zip_refs:
#  extract file into the path specified and file name is companies_sorted.csv
    zip_refs.extractall('/Users/nanayaw/Desktop/blossom/blossomenv/')
```

```python
# import pandas
import pandas as pd
# read companies CSV
ps3 = pd.read_csv('companies_sorted.csv')
# remove the null values in the columns
ps3 = ps3.dropna()
ps3.head()
```

```python
# read csv to parquet
ps3.to_parquet('free-7-million-company-dataset.parquet')
```

```python
ps3.to_json('free-7-million-company-dataset.json.gzip', compression='gzip')
```

```python
s3_client = boto3.client('s3', region_name='eu-west-1')

bucket = 'blossom-data-eng-nana'
s3_client.upload_file('free-7-million-company-dataset.json.gzip', bucket, 'free-7-million-company-dataset.json.gzip')
s3_client.upload_file('free-7-million-company-dataset.parquet', bucket, 'free-7-million-company-dataset.parquet')
s3_client.upload_file('s3_project.ipynb', bucket, 's3_project.ipynb')
```
