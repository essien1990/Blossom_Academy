```python
# import pandas
import pandas as pd
# read or import CSV file and view the data
pw = pd.read_csv('pew.csv')
pw.head()
```

```python
# rename the column name RReliggon to religion and view the data to see the change
pw = pw.rename(columns={'RReliggon':'religion',})
```

```python
pw.head()
```

```python
pw.head()
```

```python
# use melt to categorize the religion, range of income and their frequencies
melted_pw = pd.melt(pw, id_vars='religion', var_name='income_range', value_name='frequency')
melted_pw
```
